```python
import pandas as pd
from pandas.io import sql
from sqlalchemy import create_engine
```

```python
df = pd.read_csv('Stores.csv') 
mysql_host = "localhost"
mysql_port = "3306"
mysql_pwd = "sadwingsofdestiny"
mysql_schema = "cs6400_spring19"
```

```python
City = pd.read_csv('Stores.csv',usecols = ['CityName','State','Population']).drop_duplicates(["CityName",'State']).dropna()
City
```

```python
Store = pd.read_csv('Stores.csv',usecols = ['storeid','PhoneNumber','StreetAddress','CityName','State']).drop_duplicates(["storeid"]).dropna()
Store = Store.rename(columns = {'storeid':'StoreNumber'})
Store
```

```python
ActiveManager = pd.read_csv('Stores.csv',usecols = ['Email','ManagerName']).drop_duplicates(['Email']).dropna()
ActiveManager
```

```python
list(df)
```

```python
Assign = pd.read_csv('Stores.csv',usecols = ['storeid','Email']).drop_duplicates(['Email','storeid']).dropna()
Assign = Assign.rename(columns = {'storeid':'StoreNumber','Email':'ActiveManagerEmail'})
Assign
```

```python
product_csv = pd.read_csv('Products.csv')
```

```python
list(product_csv)
```

```python
Manufacturer = pd.read_csv('Products.csv',usecols = ['MfgrName','MaxDiscount']).drop_duplicates(["MfgrName"]).dropna()
Manufacturer 
```

```python
Category=pd.read_csv('Products.csv',usecols = ['CatName']).drop_duplicates(["CatName"]).dropna().reset_index(drop=True)
Category
```

```python
Product= pd.read_csv('Products.csv',usecols = ['PID', 'MfgrName', 'ProductName', 'RetailPrice']).drop_duplicates(["PID"]).dropna().reset_index(drop=True)
Product
```

```python
BelongTo = pd.read_csv('Products.csv',usecols = ['PID', 'CatName']).drop_duplicates(["PID","CatName"]).dropna().reset_index(drop=True)
BelongTo
```

```python
sales_csv = pd.read_csv('Sales.csv')
list(sales_csv)
```

```python
Day= pd.read_csv('Sales.csv',usecols = ['Date']).drop_duplicates(["Date"]).dropna().reset_index(drop=True)
Day
```

```python
Sell= pd.read_csv('Sales.csv',usecols = ['StoreNumber','Date', 'PID', 'Quantity']).drop_duplicates(["StoreNumber","Date", "PID"]).dropna().reset_index(drop=True)
Sell = Sell.rename(columns = {'Date':'SellDate'})
Sell
```

```python
saleprices = pd.read_csv('SalePrices.csv')
list(saleprices)
```

```python
OnSale= pd.read_csv('SalePrices.csv',usecols = ['Date', 'Product', 'SalePrice']).drop_duplicates(['Date', 'Product']).dropna().reset_index(drop=True)
OnSale = OnSale.rename(columns = {'Product':'PID'})
OnSale
```

```python
holiday_csv = pd.read_csv('Holidays.csv')
list(holiday_csv)
```

```python
Holiday= pd.read_csv('Holidays.csv',usecols = ['Date', 'HolidayName']).drop_duplicates(["Date"]).dropna().reset_index(drop=True)
Holiday
```

```python
engine = create_engine('mysql://root:sadwingsofdestiny@localhost/cs6400_spring19')
with engine.connect() as conn, conn.begin():
    ActiveManager.to_sql('ActiveManager',conn,if_exists='append',index=False)
    Category.to_sql('Category',conn,if_exists='append',index=False)
    City.to_sql('City',conn,if_exists='append',index=False)
    Store.to_sql('Store',conn,if_exists='append',index=False)
    Assign.to_sql('Assign',conn,if_exists='append',index=False)
    Day.to_sql('Day',conn,if_exists='append',index=False)
    Holiday.to_sql('Holiday',conn,if_exists='append',index=False)
    Manufacturer.to_sql('Manufacturer',conn,if_exists='append',index=False)
    Product.to_sql('Product',conn,if_exists='append',index=False)
    BelongTo.to_sql('BelongTo',conn,if_exists='append',index=False)
    OnSale.to_sql('OnSale',conn,if_exists='append',index=False)
    Sell.to_sql('Sell',conn,if_exists='append',index=False)
    
   
```
